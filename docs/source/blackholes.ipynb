```python
from synthpop.blackholes import accretion_history_models, accretion_history
import numpy as np
import matplotlib.pyplot as plt
from unyt import yr, Msun
```

```python
accretion_history_model = accretion_history_models.Constant(accretion_rate=1e-2 * Msun / yr, max_age=1e9 * yr)

print(accretion_history_model.final_mass.to('Msun'))
```

```python
ah = accretion_history.AccretionHistory(accretion_history_model)

ah.plot_accretion_history()
ah.plot_mass_history()
ah.plot_eddington_ratio_history()
```

## Stochastic

```python
accretion_history_model = accretion_history_models.Stochastic(max_age=2E9*yr)
ah = accretion_history.AccretionHistory(accretion_history_model)

ah.plot_accretion_history()
ah.plot_mass_history()
ah.plot_eddington_ratio_history()
```

## Constant Eddington

```python
accretion_history_model = accretion_history_models.ConstantEddington(duration=1e9 * yr, eddington_ratio=0.1)
ah = accretion_history.AccretionHistory(accretion_history_model)

ah.plot_accretion_history()
ah.plot_mass_history()
ah.plot_eddington_ratio_history()
```
